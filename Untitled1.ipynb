```python
import numpy as np
from qiskit import *
import mthree

from qiskit.test.mock.backends import FakeCasablanca
```

```python
qc = QuantumCircuit(5, 4)
qc.x(4)
qc.h(range(5))
qc.cx(range(4), 4)
qc.draw()
qc.h(range(4))
qc.barrier()
qc.measure(range(4), range(4))
qc.draw()
```

```python
backend = FakeCasablanca()
```

```python
trans_qc = transpile(qc, backend, optimization_level=3, seed_transpiler=12345)
trans_qc.draw('mpl')
```

```python
mapping = mthree.utils.final_measurement_mapping(trans_qc)
```

```python
list(mapping)
```

```python
mthree.mitigation._balanced_cal_strings(5)
```

```python
qc = QuantumCircuit(5, 4)
qc.reset(range(5))
qc.x(4)
qc.h(range(5))
qc.cx(range(4), 4)
qc.draw()
qc.h(range(4))
qc.barrier()
qc.measure(range(4), range(4))
qc.draw('mpl')
```

```python
backend = FakeCasablanca()
trans_qc = transpile(qc, backend)
trans_qc.draw('mpl')
```

```python
raw_counts = execute(trans_qc, backend).result().get_counts()
```

```python
mit = mthree.M3Mitigation(backend)
```

```python
mit.cals_from_system(list(mapping))
```

```python
quasi = mit.apply_correction(raw_counts, list(mapping))
```

```python
quasi.expval()
```
